```python
## Declaration and Assigning variables
name = "Alice"
age = 30
height = 1.65
is_student = False

## Printing variables
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
```

```python
## Naming Concentions
# Variable names should be descriptive and use lowercase letters with underscores
# They must start with a letter or an underscore and cannot be a reserved keyword and cannot contain spaces or special characters (except for underscores).
# variables are case-sensitive, meaning that "Name" and "name" would be considered different variables.
# Example of valid variable names:
first_name = "Bob"
last_name = "Smith"
total_price = 19.99
is_valid = True
# Example of invalid variable names:
# 1name = "Charlie"  # Cannot start with a number
# class = "Math"     # Cannot use reserved keywords
```

```python
## Understanding Varibale Types
# Python is dynamically typed, meaning that you don't need to declare the type of a variable when you create it. The type is inferred based on the value assigned to it.
name = "Alice"  # This is a string
age = 30        # This is an integer
height = 1.65   # This is a float
is_student = False  # This is a boolean
# You can check the type of a variable using the built-in type() function
print(type(name))  # Output: <class 'str'>
print(type(age))   # Output: <class 'int'>
print(type(height))  # Output: <class 'float'>
print(type(is_student))  # Output: <class 'bool'>
```

```python
## Type checking and conversion
# You can check the type of a variable using the type() function
# print(type(name))  # Output: <class 'str'>
# print(type(age))   # Output: <class 'int'>

## Type conversion
# You can convert between different types using built-in functions
# Convert an integer to a string
print(type(age))  # Output: <class 'int'>
age_str = str(age)
print(type(age_str))  # Output: <class 'str'>

## Convert a string to an integer
age_str = "30"
age_int = int(age_str)
print(type(age_int))  # Output: <class 'int'>
## Convert a float to an integer (this will truncate the decimal part)
height = 1.65
height_int = int(height)
print(type(height_int))  # Output: <class 'int'>
print(height_int)  # Output: 1

## Conversion doesn't always work and can raise errors if the value is not compatible with the target type
# For example, trying to convert a non-numeric string to an integer will raise a ValueError
invalid_str = "abc"

invalid_int = int(invalid_str)
# This will raise a ValueError: invalid literal for int() with base 10: 'abc'

```

```python
## Dynamic Typing
# In Python, you can reassign a variable to a different type without any issues
my_variable = "Hello"  # Initially a string
print(type(my_variable))  # Output: <class 'str'>
my_variable = 42  # Reassign to an integer
print(type(my_variable))  # Output: <class 'int'>
```

```python
## Input
# You can get user input using the input() function, which always returns a string
user_input = input("What's Your age?")
print("You entered:", user_input)
print(type(user_input))  # Output: <class 'str'>

# If you want to convert the input to a different type, you can use type conversion functions
age = int(user_input)  # Convert the input string to an integer
print("Your age is:", age)
print(type(age))  # Output: <class 'int'>
```
