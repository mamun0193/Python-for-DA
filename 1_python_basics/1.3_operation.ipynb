## Operators in Python

``Operators:`` In programming, an operator is a symbol or function that performs a specific operation on one or more operands. Operators are used to manipulate data and perform various calculations. They can be categorized into different types based on their functionality, such as arithmetic operators, logical operators, comparison operators, and assignment operators.
Here are some common types of operators:

1. Arithmetic Operators: These operators perform basic mathematical operations. Examples include:
- `+` (addition)
- `-` (subtraction)
- `*` (multiplication)
- `/` (division)
- `//` (floor division)
- `%` (modulus)
- `**` (exponentiation)
2. Logical Operators: These operators are used to combine or modify boolean values. Examples include:
   - `&&` (logical AND)
   - `||` (logical OR)
   - `!` (logical NOT)
3. Comparison Operators: These operators compare two values and return a boolean result. Examples include:
   - `==` (equal to)
   - `!=` (not equal to)
   - `>` (greater than)
   - `<` (less than)
   - `>=` (greater than or equal to)
   - `<=` (less than or equal to)
4. Assignment Operators: These operators are used to assign values to variables. Examples include:
   - `=` (simple assignment)
   - `+=` (add and assign)
   - `-=` (subtract and assign)
   - `*=` (multiply and assign)
   - `/=` (divide and assign)
   - `%=` (modulus and assign)
5. Bitwise Operators: These operators perform operations on individual bits of data. Examples include:
   - `&` (bitwise AND)
   - `|` (bitwise OR)
   - `^` (bitwise XOR)
   - `~` (bitwise NOT)
   - `<<` (left shift)
   - `>>` (right shift)
6. Ternary Operator: This operator is a shorthand for an if-else statement. It takes three operands and is represented as `? :`. For example:
   ```c
   int result = (condition) ? value_if_true : value_if_false;
   ```
   Operators play a crucial role in programming as they allow developers to perform various operations on data, manipulate values, and control the flow of the program. Understanding how to use operators effectively is essential for writing efficient and readable code.


### Arithmetic Operators

````python
## Arthemetic Operators: These operators perform basic mathematical operations. Examples include:

a=21
b=5

add_result = a + b
print("Addition:", add_result)
sub_result = a - b
print("Subtraction:", sub_result)
mul_result = a * b
print("Multiplication:", mul_result)
div_result = a / b
print("Division:", div_result)
floor_div_result = a // b
print("Floor Division:", floor_div_result)
mod_result = a % b
print("Modulus:", mod_result)
exp_result = a ** b
print("Exponentiation:", exp_result)
````

### Comparison Operators

````python
## Comparison Operators: These operators compare two values and return a boolean result. Examples include

# Equal to and not equal to
a=10
b=10
print("Equal to:", a == b) #True
print("Not equal to:", a != b) #False

# Greater than and less than
a=15
b=10
print("Greater than:", a > b) #True
print("Less than:", a < b) #False
````

### Logical Operators

````python
## Logical Operators: These operators are used to combine or modify boolean values. Examples include:

# And, Or, Not

#AND

a= True
b= False
print("Logical AND:", a and b) #False

x= True
y= True
print("Logical AND:", x and y) #True
````

````python
# OR

a= True
b= False
print("Logical OR:", a or b) #True

c= True
d= True
print("Logical OR:", c or d) #True

e= False
f= False
print("Logical OR:", e or f) #False

````

````python
# Not
a= True
print(not a) #False
````

### Assignment Operators

````python
## Assignment Operators: These operators are used to assign values to variables. Examples include:

#Simple assignment
x = 10

# Addition assignment
x += 5  # Equivalent to x = x + 5
print("Addition Assignment:", x)  # Output: 15

# Subtraction assignment
x -= 3  # Equivalent to x = x - 3
print("Subtraction Assignment:", x)  # Output: 12

# Multiplication assignment
x *= 2  # Equivalent to x = x * 2
print("Multiplication Assignment:", x)  # Output: 24

# Division assignment
x /= 4  # Equivalent to x = x / 4
print("Division Assignment:", x)  # Output: 6.0

# Modulus assignment
x %= 4  # Equivalent to x = x % 4
print("Modulus Assignment:", x)  # Output: 2.0
````

### Bitwise Operators

````python
## Bitwise Operators: These operators perform operations on individual bits of integers. Examples include:

m = 12  # 1100 in binary
n = 5   # 0101 in binary

print("Bitwise AND:", m & n)        # 4
print("Bitwise OR:", m | n)         # 13
print("Bitwise XOR:", m ^ n)        # 9
print("Bitwise NOT:", ~m)           # -13
print("Left Shift:", m << 1)        # 24
print("Right Shift:", m >> 1)       # 6
````
