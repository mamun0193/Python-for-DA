
``Data types`` — Overview

Definition:
- A data are the classififcation of data which tells the compiler or interpreter how the programmer intends to use the data. It defines the operations that can be performed on the data, the meaning of the data, and the way values of that type can be stored.

Importance:
- Enables correct operations, memory/layout assumptions, validation, serialization, performance optimizations,
    clearer APIs, and static checking.

Basic (primitive) types:
- int: integers (arbitrary precision in Python).
- float: floating-point numbers (approximate).
- bool: True/False (subclass of int).
- str: immutable Unicode text.
- NoneType: absence of a value (None).
- bytes / bytearray / memoryview: binary data and buffer views.

Core collections:
- list: ordered, mutable sequence.
- tuple: ordered, immutable sequence.
- set / frozenset: unordered unique elements (mutable/immutable).
- dict: mapping from keys to values (mutable).

Advanced / higher-level types:
- class / object: user-defined composite types with methods/attributes.
- dataclass / namedtuple / TypedDict: structured records with less boilerplate.
- enum.Enum: enumerated constants.
- typing constructs: Optional, Union, Any, Sequence, Mapping, Callable, Generic, Protocol, NewType.
- iterators / generators: lazy, on-demand sequences (yield).
- async / coroutine types: async functions, awaitables, Tasks.
- Decimal / Fraction: exact/controlled-precision numeric types.
- datetime, date, time, timedelta: temporal types.
- numpy.ndarray, pandas.Series/DataFrame: high-performance numeric/tabular types.
- memoryview / buffer protocol: zero-copy binary operations.

Key properties:
- Mutability: affects copying, sharing, thread-safety, and hashability.
- Hashability: required for dict keys / set members; immutable types are typically hashable.
- Identity vs equality: 'is' checks identity, '==' checks value equality.
- Precision & range: choose Decimal or Fraction for exactness where floats fail.
- Serialization: some types require conversion for JSON/CSV/pickle.

Practical tips:
- Use simple built-ins when sufficient; reach for dataclasses/typing for clearer APIs.
- Validate and convert external input early.
- Choose data types that match domain needs (e.g., Decimal for money, datetime for timestamps).
- Use numpy/pandas for large-scale numeric/tabular data for performance.

```python
## float type
y = 3.14
print(y, type(y))
```

```python
## Boolean type
is_valid = True
print(is_valid, type(is_valid))
```

```python
## Another type of boolean
a=10
b=20
print(a > b, type(a > b))
```
