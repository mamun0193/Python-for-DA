## Introduction
Dictionaries are a fundamental data structure in programming that allow you to store and retrieve data using key-value pairs. They are widely used in various programming languages, including Python, JavaScript, and others. In key-value pairs, keys are unique identifiers and are immutable, while values can be of any data type and can be mutable. This makes dictionaries a versatile and efficient way to manage data.

### Creating Dictionaries
In Python, dictionaries can be created using curly braces `{}` or the `dict()` constructor.
```python
# Using curly braces
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
```

```python
# Using the dict() constructor
my_dict = dict(name="Alice", age=30, city="New York")
```
### Accessing Values
You can access values in a dictionary by using their corresponding keys. If the key does not exist
, it will raise a `KeyError`. To avoid this, you can use the `get()` method, which returns `None` or a default value if the key is not found.
```python
# Accessing values using keys
print(my_dict["name"])  # Output: Alice
```
```python
# Using the get() method
print(my_dict.get("age"))  # Output: 30
```




````python
## Creating Dictionaries
empty_dict = {}
print(type(empty_dict))  # Output: <class 'dict'>
empty_dict = dict()
print(type(empty_dict))  # Output: <class 'dict'>
````

````python
student = {
    "name": "Mamun",
    "class": 3,
    "roll_number": 34,
}
print(student)  # Output: {'name': 'Mamun', 'class': 3, 'roll_number': 34}
````

````python
## ERROR
student = { "name": "Mamun", "age": 22, "name": 34, }
print(student)  # Output: {'name': 34, 'age': 22} (the last value for the duplicate key is retained)
````

````python
## Accessing Dictionary Values
new_student = {
    "name": "Mamun",
    "class": 3,
    "roll_number": 34,
}
print(new_student["name"])  # Output: Mamun
print(new_student["class"])  # Output: 3
print(new_student["roll_number"])  # Output: 34

## Accessing using get() method
print(new_student.get("name"))  # Output: Mamun
print(new_student.get("last_name"))  # Output: None (key not found)
print(new_student.get("last_name", "Not Available"))  # Output: Not Available(default value provided)
````

````python
## Modyfying Dictionary Values

student = {
    "name": "Mamun",
    "class": 3,
    "roll_number": 34,
}
student["name"] = "John"
print(student["name"])  # Output: John
````

````python
student["class"] = 4  # Update the value of an existing key
print(student)  # Output: {'name': 'John', 'class': 4, 'roll_number': 34}
student["age"] = 22  # Adding a new key-value pair
print(student)  # Output: {'name': 'John', 'class': 4, 'roll_number': 34, 'age': 22}
````

````python
## deleting Dictionary Elements
# del student["name"]  # Delete a key-value pair
# print(student)  # Output: {''class': 4, 'roll_number': 34, 'age': 22}
del student  # Delete the entire dictionary
print(student)  # Output: NameError: name 'student' is not defined
````

````python
## Dictionary Methods
student = {
    "name": "Mamun",
    "class": 3,
    "roll_number": 34,
}
keys = student.keys()  # Get all keys
print(keys)  # Output: dict_keys(['name', 'class', 'roll_number'])
values = student.values()  # Get all values
print(values)  # Output: dict_values(['Mamun', 3, 34])

items = student.items()  # Get all key-value pairs
print(items)  # Output: dict_items([('name', 'Mamun'), ('class', 3), ('roll_number', 34)])
````

````python
## copy 

student_copy = student 
print(student_copy)  # Output: {'name': 'Mamun', 'class': 3, 'roll_number': 34}
print(student)  # Output: {'name': 'Mamun', 'class': 3, 'roll_number': 34}

student['name'] = "John"  # Modify the original dictionary
print(student_copy)  # Output: {'name': 'John', 'class': 3, 'roll_number': 34}
print(student)  # Output: {'name': 'John', 'class': 3, 'roll_number': 34}

# Name changed on both student and student_copy because they refer to the same dictionary object in memory.
````

````python
## Shallow Copy
student = {
    "name": "Mamun",
    "class": 3,
    "roll_number": 34,
}
student_copy = student.copy()  # Create a shallow copy of the dictionary

student['name'] = "Alice"  # Modify the original dictionary
print(student_copy)  # Output: {'name': 'Mamun', 'class': 3, 'roll_number': 34}
print(student)  # Output: {'name': 'Alice', 'class': 3, 'roll_number': 34}


````

````python
## Itreting through a Dictionary
student = {
    "name": "Mamun",
    "class": 3,
    "roll_number": 34,
}
# interating through keys
for key in student:
    print(key)
````

````python
## iterating through values
for value in student.values():
    print(value)

## iterating through key-value pairs
for key, value in student.items():
    print(key, ":", value)
````

````python
## Nested Dictionaries

student = {
    "name": "Mamun",
    "class": 3,
    "roll_number": 34,
    "address": {
        "street": "123 Main St",
        "city": "Anytown",
        "state": "CA"
    }
}

## Access Nested Dictionary Values
print(student["address"]["city"])  # Output: Anytown
print(student["address"]["state"])  # Output: CA
````

````python
### Iterating through Nested Dictionary
for key, value in student.items():
    if isinstance(value, dict):
        print(key + ":")
        for nested_key, nested_value in value.items():
            print("  " + nested_key + ":", nested_value)
    else:
        print(key + ":", value)
````

````python
# Dictionary Comprehension

squares = {x: x**2 for x in range(1, 6)}
print (squares)  # Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
````

````python
## Conditional Dictionary Comprehension
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(even_squares)  # Output: {2: 4, 4: 16, 6: 36, 8: 64, 10: 100}
````

````python
## Practical Example: Counting Occurrences of Elements in a List

elements = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
counts = {element: elements.count(element) for element in set(elements)}
print(counts)  # Output: {'apple': 3, 'banana': 2, 'cherry': 1}
````

````python
## Merge 2 dictionaries into one 
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
merged_dict = {**dict1, **dict2}
print(merged_dict)  # Output: {'a': 1, 'b': 3, 'c': 4} (the value from dict2 overwrites the value from dict1 for the key 'b')
````
