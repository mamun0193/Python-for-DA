## Sets
Sets are built-in data structures in Python that store unordered collections of unique elements meaning the elements don't follow a specific order and duplicates are not allowed. 

### Importance of Sets
- They are useful for membership testing, removing duplicates from a sequence, and performing mathematical operations like union, intersection, and difference.
- They provide efficient methods for checking if an item is in a set, which is faster than checking in a list or tuple.
### Creating Sets
You can create a set by using curly braces `{}` or the `set()` constructor. Here are some examples:

```python
# Creating a set using curly braces
my_set = {1, 2, 3, 4, 5}

# Creating a set using the set() constructor
my_set2 = set([1, 2, 3, 4, 5])
```
### Adding Elements to a Set
You can add elements to a set using the `add()` method. If the element already exists in the set, it will not be added again.

```python
# Adding elements to a set
my_set.add(6)  # Adds 6 to the set
```
### Removing Elements from a Set
You can remove elements from a set using the `remove()` method. If the element does not exist in the set, it will raise a `KeyError`. Alternatively, you can use the `discard()` method, which does not raise an error if the element is not found.
```python
# Removing Elements from a Set
my_set.remove(3)  # Removes 3 from the set
my_set.discard(10)  # Discards 10 from the set (no error if not found)
```
### Set Operations
Sets support various mathematical operations like union, intersection, difference, and symmetric difference.

```python
# Union
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1.union(set2)  # {1, 2, 3, 4, 5}
```
```python
# Intersection
set1 = {1, 2, 3}
set2 = {3, 4, 5}
intersection_set = set1.intersection(set2)  # {3}
```
```python
# Difference
set1 = {1, 2, 3}
set2 = {3, 4, 5}
difference_set = set1.difference(set2)  # {1, 2}
```
```python
# Symmetric Difference
set1 = {1, 2, 3}
set2 = {3, 4, 5}
symmetric_diff_set = set1.symmetric_difference(set2)  # {1, 2, 4, 5}
```

### Set Methods
isdisjoint(): Returns True if two sets have no elements in common.
```python
is_disjoint = set1.isdisjoint(set2)  # Output: False (since they share the element 3)
```
issubset(): Returns True if all elements of one set are in another set.
```python
is_subset = set1.issubset(set2)  # Output: False (since not all elements of set1 are in set2)
```
issuperset(): Returns True if all elements of one set are in another set.
```python
is_superset = set1.issuperset(set2)  # Output: False (since not all elements of set2 are in set1)
```

#### Conclusion
Sets are a powerful and efficient data structure in Python that allow for the storage of unique elements and provide various operations for manipulating these elements. They are particularly useful in scenarios where you need to ensure that no duplicates are present and when performing mathematical set operations.

````python
## Create a Set

my_set = {1, 2, 3, 4, 5}
print(my_set)  # Output: {1, 2, 3, 4, 5}
print(type(my_set))  # Output: <class 'set'>
````

````python
my_empty_set = set()
print(my_empty_set)  # Output: set()
print(type(my_empty_set))  # Output: <class 'set'>
````

````python
my_unique_set = set([1, 1, 2, 2, 3, 3])
print(my_unique_set)  # Output: {1, 2, 3}: duplicate values are removed here
print(type(my_unique_set))  # Output: <class 'set'>
````

````python
## Adding elements to a Set

my_set.add(6)
print(my_set)  # Output: {1, 2, 3, 4, 5, 6}
````

````python
## Removing elements from a Set
my_set.remove(3)
print(my_set)  # Output: {1, 2, 4, 5, 6}
````

````python
my_set.discard(10)  # Does not raise an error if the element is not present
````

````python
## Pop method
removed_element = my_set.pop()  # Removes and returns an arbitrary element
print(f"Removed element: {removed_element}")
print(f"Set after pop: {my_set}")
````

````python
## Clearing a Set
my_set.clear()
print(my_set)  # Output: set()
````

````python
## Set Membership Testing
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)  # Output: True
print(6 in my_set)  # Output: False
````

````python
## Mathematical Set Operations
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
### Union
union_set = set_a.union(set_b)
print(union_set)  # Output: {1, 2, 3, 4, 5, 6, 7, 8}
### Intersection
intersection_set = set_a.intersection(set_b)
print(intersection_set)  # Output: {4, 5}
### Difference
difference_set = set_a.difference(set_b)
print(difference_set)  # Output: {1, 2, 3} it shows the elements that are in set_a but not in set_b
### Symmetric Difference
symmetric_difference_set = set_a.symmetric_difference(set_b)
print(symmetric_difference_set)  # Output: {1, 2, 3, 6, 7, 8}
````

````python
## Set Methods 
set1 = {1, 2, 3}
set2 = {3, 4, 5}
### isdisjoint
is_disjoint = set1.isdisjoint(set2)
print(is_disjoint)  # Output: False, because they have a common element (3
### issubset
is_subset = set1.issubset(set2)
print(is_subset)  # Output: False, because not all elements of set1 are in set2
### issuperset
is_superset = set1.issuperset(set2)
print(is_superset)  # Output: False, because not all elements of set2 are in set1

set3 = {1, 2}
set4 = {1, 2, 3, 4}
is_subset_true = set3.issubset(set4)
print(is_subset_true)  # Output: True, because all elements of set3 are in set4
is_superset_true = set4.issuperset(set3)
print(is_superset_true)  # Output: True, because all elements of set3 are in set4

set5 = {1, 2, 3}
set6 = {4, 5, 6}
is_disjoint_true = set5.isdisjoint(set6)
print(is_disjoint_true)  # Output: True, because they have no common elements
````

````python
### Counting Unique words in a sentence

sentence = "This is a sample sentence with some sample words"
words = sentence.split()
unique_words = set(words)
print(f"Unique words: {unique_words}")
print(f"Count of unique words: {len(unique_words)}")
````
