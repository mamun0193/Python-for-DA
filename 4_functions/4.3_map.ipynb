## Map function

The `map()` function in Python applies a given function to each item of an iterable (like a list or tuple) and returns a map object (which is an iterator). You can convert the map object to a list or other iterable if needed.


```python
def square(x):
    return x * x
    
print(square(2))
```

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(square, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

```python
## Lambda function with map

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

```python
## Map multiple iterables

numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

added_numbers = list(map(lambda x, y: x + y, numbers1, numbers2))
print(added_numbers)  # Output: [5, 7, 9]
```

```python
## Map to convert list of strings to integers

string_numbers = ['1', '2', '3', '4', '5']
integer_numbers = list(map(int, string_numbers))
print(integer_numbers)  # Output: [1, 2, 3, 4, 5]
```

```python
## To uppercase all strings in a list

strings = ['hello', 'world', 'python']
uppercase_strings = list(map(str.upper, strings))
print(uppercase_strings)  # Output: ['HELLO', 'WORLD', 'PYTHON']
```

```python
def get_name(person):
    return person['name']
people = [{'name': 'Alice', 'age': 30}, {'name': 'Bob', 'age': 25}, {'name': 'Charlie', 'age': 35}]
names = list(map(get_name, people))
print(names)  # Output: ['Alice', 'Bob', 'Charlie']
```

### Conclusion
The map() function is a powerful tool for applying transformations to iterables in a concise and readable manner. It can be used with both named functions and lambda functions, and it can handle multiple iterables as well.
