## Lambda Functions
Lambda functions, also known as anonymous functions, are small, unnamed functions defined using the `lambda` keyword in Python. They can take any number of arguments but can only have one expression. The expression is evaluated and returned when the function is called.They are commonly used for short, simple operations or as arguments to higher-order functions like `map()`, `filter()`, and `reduce()`.
Syntax:
```python
lambda arguments: expression
```

````python
additon = lambda x, y: x + y
print(additon(5, 3))  # Output: 8
````

````python
even= lambda x: x % 2 == 0
print(even(4))  # Output: True
````

````python
additon1 = lambda a,b,c : a+b+c
print(additon1(5, 3, 2))  # Output: 10
````

````python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9,
````
